```python
data={"YEAR/MONTH":[],"MONTHLY REVENUE":[]}
df2=pd.DataFrame(data)
df2.to_excel("monthly_revenue.xls",sheet_name="Sheet 1")
df2
```

```python
df2["YEAR/MONTH"]=df["Order Date"].dt.strftime("%Y-%m")
df2["MONTHLY REVENUE"]=df['Sales']
df2=df2.groupby("YEAR/MONTH")["MONTHLY REVENUE"].sum()
df2
```

```python
df2.plot(x="YEAR/MONTH",y="Order Date",color="g")
plt.title("Monthly Revenue")
plt.xlabel("YEAR/MONTH")
plt.ylabel("MONTHLY REVENUE")
plt.rc("figure",figsize=(30,15))
plt.show()
```

```python
df2.describe()
```
